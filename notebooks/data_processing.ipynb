```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
import seaborn as sns
```

```python
# 시각화 스타일
sns.set(style='whitegrid')
```

```python
# 1. 데이터 로드
df = pd.read_csv('../data/HR-Employee-Attrition.csv')

df.head()
```

```python
# 2. 데이터 이해
df.shape
df.columns.size
df.info()
df.describe()
```

```python
# 결측값 확인
df.isnull().sum()
```

```python
# 3. 데이터 전처리
# Attrition 숫자 변환
df['Attrition'] = df['Attrition'].map({'Yes': 1, 'No': 0})
```

```python
# 나이 그룹 생성 (세대 분석 핵심)
# 나이의 최소값, 최대값 확인
df['Age'].min()
df['Age'].max()

bins = [18, 30, 40, 50, 60]
labels = ['20s', '30s', '40s', '50s']

df['AgeGroup'] = pd.cut(df['Age'], bins=bins, labels=labels, right=False)
```

```python
# 4. 전체 이탈 현황
attrition_rate = df['Attrition'].mean()
print(f"전체 이탈률: {attrition_rate:.2%}")
```

```python
sns.countplot(x='Attrition', data=df)
plt.title('Attrition Distribution')
plt.savefig('../images/attrition_distribution.png')
plt.show()
```
