```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
import seaborn as sns
```

```python
# 시각화 스타일
sns.set(style='whitegrid')
```

```python
# 1. 데이터 로드
df = pd.read_csv('../data/HR-Employee-Attrition.csv')

df.head()
```

```python
# 2. 데이터 이해
df.shape
df.columns.size
df.info()
df.describe()
```

```python
# 결측값 확인
df.isnull().sum()
```

```python
# 3. 데이터 전처리
# Attrition 숫자 변환
df['Attrition'] = df['Attrition'].map({'Yes': 1, 'No': 0})
```

```python
# 4. 전체 이탈 현황
attrition_rate = df['Attrition'].mean()
print(f"전체 이탈률: {attrition_rate:.2%}")
```

```python
# 전체 분포 비교
sns.countplot(x='Attrition', data=df)
plt.title('Attrition Distribution')
plt.show()
```

```python
# 5. 핵심 변수 분석
# 추가 근무에 따른 이탈율 분석
df. groupby('OverTime')['Attrition'].mean().round(2)

# 전체 중 추가 근무를 한 인원이 이탈하는 확률이 하지 않은 직원들 보다 3배 더 높음.
```

```python
# 추가 근무에 따른 이탈율 시각화

sns.barplot(x='OverTime' , y='Attrition', data= df, ci= None)
plt.title('Attrition Rate by OverTime')
plt.ylabel('Attrition Rate')
plt.xlabel('OverTime')
plt.show()

# 추가 근무가 많을 수록 이탈이 증가함 -> 워라밸 문제로 이어짐
```

```python
# 근무 만족도에 따른 이탈율 분석

df.groupby('JobSatisfaction')['Attrition'].mean().round(2)
```

```python
# 근무 만족도에 따른 이탈율 시각화 비교

sns.barplot(x='JobSatisfaction' , y='Attrition', data=df, ci= None)
plt.title('Attrition Rate by JobSatisfaction')
plt.show()

# 근무 만족도가 낮을수록 이탈율 증가
```

```python
# 연봉에 따른 이탈율 분석

df.groupby('Attrition')['MonthlyIncome'].mean().round(2)
```

```python
# 연봉에 따른 이탈율 비교

sns.boxplot(x='Attrition', y='MonthlyIncome', data=df, showfliers = False)
plt.title('Attrition by Income')
plt.show()

# 낮은 연봉으로 인해 이탈 가능성 높음.
```

```python
# 근속 연수 확인
df.groupby('YearsAtCompany').size()
```

```python
# 근속 연수에 따른 이탈율 분석

yearbins = [0, 5, 10, 20, 50]
yearlabels = ['0-5', '5-10', '10-20', '20+']

df['YearsGroup'] = pd.cut(df['YearsAtCompany'], bins=yearbins , labels=yearlabels, right=False)

```

```python
# 그룹 별 실제 이탈한 인원이 몇 명인지 확인
years_analysis = df.groupby('YearsGroup').agg( 
    Total_count = ('Attrition', 'count'),
    Attrition_count = ('Attrition', 'sum')
    )

# 숫자형 유지
years_analysis['Attrition_rate(%)'] = ((years_analysis['Attrition_count'] / years_analysis['Total_count']) * 100).round(2)
years_analysis
# years_analysis['Total_Count'].sum()
```

```python
# 근속 연수 그룹별 비교

sns.barplot(x=years_analysis.index, y= years_analysis['Attrition_rate(%)'], ci=None)

plt.title('Attrition Rate by Years at Company')
plt.ylabel('Attrition Rate')
plt.xlabel('Years Group')

plt.show()
```

```python
# 세대(나이) 분석
# 나이 그룹 생성 (세대 분석 핵심)
# 나이의 최소값, 최대값 확인
df['Age'].min()
df['Age'].max()

age_bins = [18, 30, 40, 50, 65]
age_labels = ['20s', '30s', '40s', '50s']

df['AgeGroup'] = pd.cut(df['Age'], bins=age_bins, labels=age_labels, right=False)
```

```python
age_analysis = df.groupby('AgeGroup').agg(
    Total_count = ('Attrition' , 'count'), 
    Attrition_count = ('Attrition' , 'sum'), 
)
# 이탈율 백분율 변환
age_analysis['Attrition_rate(%)'] = ((age_analysis['Attrition_count'] / age_analysis['Total_count']) * 100).round(2)
age_analysis
```

```python
sns.barplot(x=age_analysis.index, y=age_analysis['Attrition_rate(%)'], ci=None)

plt.title('Attrition Rate by Age Group')
plt.ylabel('Attrition Rate(%)')
plt.xlabel('Age Group')

plt.show()
```

```python
# 부서별 이탈율 분석

dept_analysis = df.groupby('Department').agg(
    Total_count = ('Attrition', 'count'),
    Attrition_count = ('Attrition', 'sum')
)

dept_analysis['Attrition_rate(%)'] = ((dept_analysis['Attrition_count'] / dept_analysis['Total_count']) * 100).round(2)
dept_analysis_sorted = dept_analysis.sort_values(by='Attrition_rate(%)', ascending=False)
dept_analysis_sorted
```

```python
### 부서별 이탈율 정렬 

sns.barplot(x=dept_analysis_sorted.index, y=dept_analysis_sorted['Attrition_rate(%)'], ci=None)

plt.title('Attrition Rate by Department')
plt.ylabel('Attrition Rate(%)')
plt.xlabel('Department')

plt.xticks(rotation=30)
plt.show()
```

```python
# 6. 핵심 교차 분석
## 근속 연수 + 추가 근무에 따른 이탈율

cross_1 = df.groupby(['YearsGroup', 'OverTime']).agg(
    Total_count = ('Attrition' , 'count'),
    Attrition_count = ('Attrition' , 'sum')
)

cross_1['Attrition_rate(%)'] = ((cross_1['Attrition_count'] / cross_1['Total_count']) * 100).round(2)
cross_1
```

```python
### 근속 연수와 추가 근무 대비 그룹별 이탈율 비교

pivot_1 = cross_1['Attrition_rate(%)'].unstack()

sns.heatmap(pivot_1, annot=True, fmt=".1f", cmap='coolwarm')

plt.title('Attrition Rate by Years & OverTime')

plt.show()
```

```python
## 나이 + 추가 근무에 따른 이탈율

cross_2 = df.groupby(['AgeGroup', 'OverTime']).agg(
    Total_count = ('Attrition' , 'count'),
    Attrition_count = ('Attrition' , 'sum')
)

cross_2['Attrition_rate(%)'] = ((cross_2['Attrition_count'] / cross_2['Total_count']) * 100).round(2)
cross_2
```

```python
pivot_2 = cross_2['Attrition_rate(%)'].unstack()

sns.heatmap(pivot_2, annot=True, fmt=".1f", cmap='coolwarm')

plt.title('Attrition Rate by Age & OverTime')

plt.show()
```
